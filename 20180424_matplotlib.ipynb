```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np

x = np.linspace(0,2.5,11)
y = x**0.5
```

```python
plt.plot(x,y)
plt.xlabel('Valores Numericos')
plt.ylabel('Raiz dos valores')
plt.title('Calculo de raiz quadrada')
```
